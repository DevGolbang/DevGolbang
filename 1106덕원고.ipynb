```python
import numpy as np

initializeTurtle()

goto(350,200)

r = 100

for i in range(100):
  angle = (2*np.pi)/100*i
  x = r * np.cos(angle)
  y = r * np.sin(angle)
  goto(x+350,y+200)
```
